```python
from sklearn.preprocessing import LabelEncoder
data['Sex']=LabelEncoder().fit_transform(data['Sex'])
```

```python
target = 'CGPA'

X = data.drop(target,axis=1) 
#y will be the target variable
y = data[target]
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

```

```python
model.score(X_test, y_test)
```

```python
model.score(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE: {rmse}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming train_predictions is the predicted CGPA values for x_train_preprocessed

# Plotting the distribution of actual CGPA (y_train) with blue color
sns.distplot(y_test, label='Actual CGPA', color='blue')

# Plotting the distribution of predicted CGPA (y_train_pred) with orange color
sns.distplot(y_pred, label='Predicted CGPA', color='orange')

# Adding labels and title
plt.xlabel('CGPA')
plt.ylabel('Density')
plt.title('Distribution of Actual vs. Predicted CGPA')
plt.legend()

# Displaying the plot
plt.show()
```
